```python
df = df[df['job'] != 'unknown']
df = df[df['marital'] != 'unknown']
df = df[df['education'] != 'unknown']
df = df[df['housing'] != 'unknown']
```

```python
df["job"]=df["job"].replace({"admin.":0,"blue-collar":1,"technician":2,"services":3,"management":4,
                            "retired":5,"entrepreneur":6,"self-employed":7,"housemaid":8,"unemployed":9,"student":10})
df["marital"]=df["marital"].replace({"married":0,"single":1,"divorced":2})
df["education"]=df["education"].replace({"university.degree":0,"high.school":1,"basic.9y":2,"professional.course":3,"basic.4y":4,"basic.6y":5,"illiterate":6})
df["housing"]=df["housing"].replace({"yes":1,"no":0})
df["loan"]=df["loan"].replace({"yes":1,"no":0})
df["contact"]=df["contact"].replace({"cellular":0,"telephone":1})
df["month"]=df["month"].replace({"may":0,"jul":1,"aug":2,"jun":3,"nov":4,"apr":5,"oct":6,"sep":7,"mar":8,"dec":9})
df["day_of_week"]=df["day_of_week"].replace({"thu":0,"mon":1,"wed":2,"tue":3,"fri":4})
df["y"]=df["y"].replace({"yes":1,"no":0})
df["poutcome"]=df["poutcome"].replace({"nonexistent":0,"failure":1,"success":2})
```

```python
from sklearn.preprocessing import MinMaxScaler
col=["age","duration","pdays","cons.price.idx","cons.conf.idx","euribor3m","nr.employed"]
scaler = MinMaxScaler()
df[col]=scaler.fit_transform(df[col])
```

```python
X=df.drop(columns="y",axis=1)
Y=df["y"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print("Training set shapes - X: {}, y: {}".format(X_train.shape, y_train.shape))
print("Testing set shapes - X: {}, y: {}".format(X_test.shape, y_test.shape))
```

```python
param_dist = {
    'n_estimators': randint(50, 200),
    'learning_rate': [0.1, 0.01, 0.001],
    'max_depth': randint(3, 10)
}
gb=GradientBoostingClassifier()
random_search = RandomizedSearchCV(gb, param_distributions=param_dist, n_iter=10, cv=5, random_state=42)

random_search.fit(X_train, y_train)
```

```python
best_params = random_search.best_params_
best_score = random_search.best_score_

print("Best Parameters:", best_params)
print("==================================================")
print("Best Score:", best_score)
```

```python
best_gb=GradientBoostingClassifier(learning_rate=0.1, max_depth=6, n_estimators=107)
best_gb.fit(X_train,y_train)
```

```python
feat_importances = pd.Series(best_gb.feature_importances_, index = X.columns)
feat_importances.nlargest(len(df.columns)).plot(kind = 'barh',color="indigo",figsize=(10,8))

plt.xlabel('Feature Importances')
plt.ylabel('Features Name')
plt.show()
```

```python
feat_importances.sort_values(ascending =True)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
pred=best_gb.predict(X_test)
accuracy = accuracy_score(y_test,pred)
print("Accuracy:", accuracy)
```

```python
cf=confusion_matrix(y_test,pred)
label_name=["No","Yes"]
plt.figure(figsize=(10, 8))
sns.heatmap(cf, annot=True, fmt="d", cmap="PRGn", xticklabels=label_name, yticklabels=label_name)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
print(classification_report(y_test, pred,target_names=label_name))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score


y_probs = best_gb.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
plt.figure(figsize=(10,8))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_score(y_test, y_probs))
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, auc

precision, recall, _ = precision_recall_curve(y_test, y_probs)
pr_auc = auc(recall, precision)
plt.figure(figsize=(10,8))
plt.plot(recall, precision, color='blue', lw=2, label=f'Precision-Recall Curve (AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.legend(loc="lower left")
plt.grid(True)
plt.show()
```
